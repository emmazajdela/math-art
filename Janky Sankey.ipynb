```python
import json
import plotly.graph_objects as go

# Opening JSON file
f = open('source.json')
source = json.load(f)

f = open('target.json')
target = json.load(f)

value = [1 for x in target]

link = dict(source = source, target = target,value=value)
data = go.Sankey(link = link)

fig = go.Figure(data)

fig.show()
```

```python
# Kaleido needed to run the next line
fig.write_image("JankySankey.png")
```
